##Gráficos

```python
# Grágico de Caixa ("Boxplot")
dataset.boxplot(column='idade')
```

```python
# Histograma com as idades dos acessos realizados
dataset['idade'].hist(bins=50, color='green')
```

```python
# Histograma com as idades dos acessos realizados com idade < 80 anos
dataset[dataset['idade'] < 80]['idade'].hist(bins=30, color='red') # < 80
```

```python
# Eliminando os Outliers
# Histograma de contagem total dos acessos apenas para faixa etária de adultos (> 25 anos) E do sexo masculino
dataset[(dataset['idade'] >= 18) & (dataset['idade'] <= 80) & (dataset['sexo'] == 'M')]['idade'].hist(bins=30, color='red')
```

```python
# Eliminando os Outliers
# Histograma de contagem total dos acessos apenas para faixa etária de adultos (> 25 anos) E do sexo masculino
dataset[(dataset['idade'] >= 18) & (dataset['idade'] <= 80) & (dataset['sexo'] == 'F')]['idade'].hist(bins=30, color='red')
```
