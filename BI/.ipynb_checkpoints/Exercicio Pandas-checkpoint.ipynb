```python
#Exercicio 1 - Criando uma Series vazia

import pandas as pd
series_vazia = pd.Series()
series_vazia
```

```python
#Exercicio 2 - Criando uma Series a partir de uma lista

import pandas as pd
lista_frutas = ['Maçã', 'Banana', 'Morango', 'Uva', 'Abacaxi']
series_frutas = pd.Series(lista_frutas)
series_frutas
```

```python
#Exercicio 3 - Criando uma Series a partir de um array NumPy

import pandas as pd
import numpy as np
dados = np.array([10, 20, 30, 40])
series_customizada = pd.Series(dados, index=['A', 'B', 'C', 'D'])
series_customizada
```

```python
#Exercicio 4 - Criando uma Series a partir de um dicionário

import pandas as pd
dados = {'jan': 100, 'fev': 200}
series_meses = pd.Series(dados, index=['jan', 'fev', 'mar'])
series_meses
```

```python
#Exercicio 5 - Criando um DataFrame vazio

import pandas as pd
df = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)
df
```
