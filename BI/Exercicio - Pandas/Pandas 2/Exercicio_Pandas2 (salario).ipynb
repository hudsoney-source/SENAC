```python
import pandas as pd

dados = {
    'UF': ['MS', 'MA', 'PA', 'RR', 'MT', 'PB', 'RN', 'AP', 'DF', 'SE', 'TO', 'PE', 'AL', 'ES', 'AM', 'SP', 'AC', 'GO', 'RO', 'RS', 'SC', 'PI', 'BA', 'CE', 'PR', 'MG'],
    'Salario': [13007.12, 8452.03, 8289.86, 7700.47, 7343.44, 6944.09, 6814.88, 6600.98, 6427.71, 6176.76, 6130.70, 5841.91, 5767.15, 5685.97, 5631.18, 5565.00, 5370.35, 5160.49, 5118.41, 5111.05, 5026.80, 4984.17, 4965.24, 4961.73, 4920.55, 4867.97],
    'Regiao': ['Centro-Oeste', 'Nordeste', 'Norte', 'Norte', 'Centro-Oeste', 'Nordeste', 'Nordeste', 'Norte', 'Centro-Oeste', 'Nordeste', 'Norte', 'Nordeste', 'Nordeste', 'Sudeste', 'Norte', 'Sudeste', 'Norte', 'Centro-Oeste', 'Norte', 'Sul', 'Sul', 'Nordeste', 'Nordeste', 'Nordeste', 'Sul', 'Sudeste']
}
df = pd.DataFrame(dados)

print(f"Menor salário: {df.min()['UF']} (R${df['Salario'].min()})")
print(f"Maior salário: {df.max()['UF']} (R${df['Salario'].max()})")

df_rank = df.sort_values('Salario', ascending=False).reset_index(drop=True)
posicao_go = df_rank[df_rank['UF'] == 'GO'].index[0] + 1
print(f"Goiás está na {posicao_go}ª posição")

print(f"Média salarial geral: R${df['Salario'].mean():.2f}")

resumo_regiao = df.groupby('Regiao')['Salario'].agg(['mean', 'max', 'min'])
print("\nResumo por Região:")
print(resumo_regiao)

print(f"\nRegião com MAIOR média: {resumo_regiao['mean'].idxmax()}")
print(f"Região com MENOR média: {resumo_regiao['mean'].idxmin()}")
```
