```python
import pandas as pd
from typing import Dict, Union, List

class AnaliseSalarial:
    def __init__(self, dados: Dict[str, List[Union[str, float]]]):
        """Inicializa o DataFrame e valida a estrutura dos dados."""
        self.df = pd.DataFrame(dados)
        self._preparar_dados()

    def _preparar_dados(self) -> None:
        """Ordena os dados por salário para facilitar rankings."""
        self.df = self.df.sort_values(by='Salario', ascending=False).reset_index(drop=True)

    def obter_extremos(self) -> Dict[str, dict]:
        """Retorna os estados com maior e menor salário usando idxmax/idxmin."""
        idx_max = self.df['Salario'].idxmax()
        idx_min = self.df['Salario'].idxmin()
        
        return {
            "maior": self.df.loc[idx_max].to_dict(),
            "menor": self.df.loc[idx_min].to_dict()
        }

    def ranking_estado(self, uf: str) -> int:
        """Retorna a posição de um estado no ranking (1-based)."""
        try:
            return self.df[self.df['UF'] == uf.upper()].index[0] + 1
        except IndexError:
            raise ValueError(f"Estado '{uf}' não encontrado nos dados.")

    def resumo_por_regiao(self) -> pd.DataFrame:
        """Gera estatísticas descritivas agrupadas por região."""
        return self.df.groupby('Regiao')['Salario'].agg(
            Media='mean',
            Maximo='max',
            Minimo='min',
            Contagem='count'
        ).sort_values(by='Media', ascending=False)

# --- Execução Profissional ---

if __name__ == "__main__":
    dados_brutos = {
        'UF': ['MS', 'MA', 'PA', 'RR', 'MT', 'PB', 'RN', 'AP', 'DF', 'SE', 'TO', 'PE', 'AL', 'ES', 'AM', 'SP', 'AC', 'GO', 'RO', 'RS', 'SC', 'PI', 'BA', 'CE', 'PR', 'MG'],
        'Salario': [13007.12, 8452.03, 8289.86, 7700.47, 7343.44, 6944.09, 6814.88, 6600.98, 6427.71, 6176.76, 6130.70, 5841.91, 5767.15, 5685.97, 5631.18, 5565.00, 5370.35, 5160.49, 5118.41, 5111.05, 5026.80, 4984.17, 4965.24, 4961.73, 4920.55, 4867.97],
        'Regiao': ['Centro-Oeste', 'Nordeste', 'Norte', 'Norte', 'Centro-Oeste', 'Nordeste', 'Nordeste', 'Norte', 'Centro-Oeste', 'Nordeste', 'Norte', 'Nordeste', 'Nordeste', 'Sudeste', 'Norte', 'Sudeste', 'Norte', 'Centro-Oeste', 'Norte', 'Sul', 'Sul', 'Nordeste', 'Nordeste', 'Nordeste', 'Sul', 'Sudeste']
    }

    try:
        analise = AnaliseSalarial(dados_brutos)
        
        # 1. Extremos
        extremos = analise.obter_extremos()
        print(f"🚀 Maior Salário: {extremos['maior']['UF']} - R${extremos['maior']['Salario']:.2f}")
        print(f"📉 Menor Salário: {extremos['menor']['UF']} - R${extremos['menor']['Salario']:.2f}")

        # 2. Ranking Específico
        estado_alvo = "GO"
        print(f"📍 {estado_alvo} está na {analise.ranking_estado(estado_alvo)}ª posição no ranking nacional.")

        # 3. Relatório Regional
        print("\n--- RELATÓRIO POR REGIÃO ---")
        print(analise.resumo_por_regiao())

    except Exception as e:
        print(f"Erro ao processar análise: {e}")
```
