```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb

# importar e criar o dataframe
url = 'https://raw.githubusercontent.com/Ujeverson/datasets/main/Salary_Data.csv'
wage = pd.read_csv(url)
wage.head()
```

```python
x = wage['YearsExperience']
y = wage['Salary']
```

```python
plt.scatter(x, y, color='red')
```

```python
# somatório dos valores de x
xs = x.sum()
xs
```

```python
# somatório dos valores de y
ys = y.sum()
ys
```

```python
# produto entre valores de x e y
xy = x * y
xy
```

```python
# somatório do produto entre valores de x e y
xys = (x * y).sum()
xys
```

```python
# x ao quadrado
x2 = x**2
x2
```

```python
# soma dos valores de x ao quadrado
x2s = (x**2).sum()
x2s
```

```python
# quantidade de linhas do dataframe
n = len(wage)
n
```

```python
# determinando o valor de a
a = (n * xys - xs * ys) / (n * x2s - (xs)**2)
print(a)
print(round(a, 2))
```

```python
# determinando o valor de b
b = (ys - a * xs) / n
print(b)
print(round(b, 2))
```

```python
# determinando o valor de a (inclinação)
a = (n * xys - xs * ys) / (n * x2s - (xs)**2)
print(f"Valor de a: {a}")
```

```python
# determinando o valor de b (intercepto)
b = (ys - a * xs) / n
print(f"Valor de b: {b}")
```

```python
# Predição: 5 anos de experiência
xyears = 5
ysalary = a * xyears + b
```

```python
# resultado do salário para xyears
print(f"Salário estimado para {xyears} anos: {ysalary}")
```

```python
# criando uma lista vazia para armazenar
# as previsões de todos os valores de x
# do nosso dataframe
previsoes = []
```

```python
# loop para calcular todos os valores do
# dataset e inserir em 'previsoes'.
for elemento in x:
    ypred = a * elemento + b
    previsoes.append(ypred)
```

```python
# imprimindo todas os salários
y-previsoes
```

```python
# Inserir as previsões no DataFrame
wage['predicoes'] = previsoes
```

```python
# Visualizar a Regressão Linear
plt.scatter(x, y, color = 'purple')
plt.plot(x, wage['predicoes'], color = 'green')
```
